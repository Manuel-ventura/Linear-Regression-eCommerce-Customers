# Linear Regression

Linear Regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. The goal is to establish a linear equation that best predicts the dependent variable based on the independent variables.

# Objectives of the Project

1. **Understand the relationship:** Identify how different independent variables affect the dependent variable.

2. **Prediction:** Use the established model to make predictions about future values of the dependent variable.

3. **Interpretation:** Interpret the coefficients to understand the impact of each independent variable on the dependent variable.

## Expected Findings

1. **Significant predictors:** Determine which independent variables significantly influence the dependent variable.

2. **Model accuracy:** Assess how well the linear regression model predicts the dependent variable using metrics like R-squared and Mean Squared Error (MSE).

```python
from sklearn.model_selection import train_test_split
```

```python
# Defining Features and Target Variable
X = df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
#  Splitting the Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### This code splits the data into training and testing sets. The test_size=0.3 parameter indicates that 30% of the data will be used for testing, and the random_state=101 parameter ensures reproducibility.

## Training the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

### This section imports the LinearRegression class from sklearn.linear_model, creates an instance of the linear regression model, and fits it to the training data.

```python
# Model Coefficients
lm.coef_
```

### This line prints the coefficients of the linear regression model, which represent the impact of each feature on the target variable.

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coef'])
print(cdf)
```

## Predictions

```python
# Making Predictions
predictions = lm.predict(X_test)

#This code uses the trained model to make predictions on the test set.
```

```python
predictions
```

## Visualizing Predictions vs Actual Values

```python
sns.scatterplot(x=predictions, y=y_test)
plt.xlabel("Predictions")
plt.title("Evaluation of our LM model")
```

## Evaluation of a Linear Regression Model (LM) by Comparing Model Predictions with Actual Values (y_test)

### Linear Relationship:

The points in the plot follow an almost perfect diagonal line, suggesting that the models predictions are very close to the actual values.
This indicates that the linear regression model has excellent performance and the predictions are highly accurate.

### Point Distribution:

The points are uniformly distributed along the diagonal, showing no significant bias in the predictions. The model can accurately predict both low and high values with similar precision.

### Prediction Errors:

The closeness of the points to the diagonal line indicates very small prediction errors. The variability is minimal, reinforcing the models accuracy.

### Data Range:

The predictions and actual values cover a wide range (approximately 300 to 700), suggesting that the model works well for a variety of Yearly Amount Spent values.

The plot indicates that the linear regression model has excellent performance, with predictions very close to the actual values. The almost perfect relationship between the predictions and the actual values shows that the model is very precise and reliable for this dataset.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
# Evaluating the Model
print("Mean Absolute Error: ", mean_absolute_error(y_test, predictions))
print("Mean Squared Error: ", mean_squared_error(y_test, predictions))
print("RMSE: ", math.sqrt(mean_squared_error(y_test, predictions)))
```

### This code evaluates the model's performance using various metrics, including Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared score.

## Residuals

```python
residuals = y_test - predictions
```

```python
sns.displot(residuals, bins=20, kde=True)
```

# Histogram with a Superimposed Density Curve, Representing the Distribution of Residuals from a Regression Model

The graph has the following characteristics:

- **X-Axis:** Represents the residuals ("Residuals"), which are the differences between the actual values (y_test) and the model's predictions (predictions).
- **Y-Axis:** Represents the count or frequency of the residuals in each bin of the histogram.
- **Residual Distribution:** The residuals appear to be approximately normally distributed, centered around 0. This indicates that the model does not have a significant systematic bias in the predictions.
- **Shape of the Distribution:** The symmetrical and roughly bell-shaped distribution suggests that the residuals follow a normal distribution, which is a good indication that the regression model fits the data adequately.

## Conclusions:

- **Normality of Residuals:** The approximately normal distribution of the residuals suggests that the regression model is appropriate and the assumptions of normality of the errors may be satisfied.
- **Absence of Bias:** The centrality of the distribution around 0 indicates that there is no systematic bias in the model's predictions.


```python
import pylab
import scipy.stats as stats

stats.probplot(residuals, dist="norm", plot=pylab)
pylab.show()
```

## Analysis of the Probability Plot

### Observation:

- **Linearity**: The data points (blue dots) are generally following a straight line (red line), which suggests that the residuals are approximately normally distributed.
- **Outliers**: There are some noticeable deviations from the line at both ends, indicating the presence of outliers or heavy tails in the distribution of residuals.
- **Symmetry**: The plot shows a roughly symmetric pattern around the red line, further indicating that the residuals are fairly normal, though the extreme values show some deviation.

### Conclusion:
The Q-Q plot suggests that the residuals of the regression model are approximately normally distributed, with some deviations at the tails. This means that the assumptions of normality for the residuals hold reasonably well, though the presence of outliers should be further investigated.
